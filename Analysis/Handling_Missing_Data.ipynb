#INTERPOLATE

**interpolate()**

Uses different methods to interpolate the values

ex: Col1 = 1,2,3,4,5,6,**nan**,8,9
interpolate will automatically convert the nan to 7

linearly distributed methods
5, nan, 6 -> 5, 5.5, 6 (using average)

```python
df.balcony.head(12)
```

```python
df.balcony.interpolate(method='linear').head(12)
```

```python
df.balcony.interpolate(method='linear', limit_direction='forward').head(12)
```

limit = 2, 2 of n consecutive null values will be filled

limit_direction = forward, successing nan values to apply the limit

```python
df.balcony.head(12)
```

```python
df.balcony.interpolate(method='linear', limit=1).head(12)
```

*Only the first Nan of consecutive nan is filled*
