#**Different PATTERNS for Bars**



##**var = plt.bar(.........)**

##vars[i]**.set_hatch( *pattern*)**

```python
l = ['A', 'B', 'C']
vals = [3, 5, 4]

bars = plt.bar(l, vals, color='#CCCCFF')

bars[0].set_hatch('/')

plt.show()
```

```python
l = ['A', 'B', 'C', 'D']
vals = [3, 5, 4, 2]

bars = plt.bar(l, vals, color='#CCCCFF')

bars[0].set_hatch('o')
bars[1].set_hatch('+')
bars[2].set_hatch('|')
bars[3].set_hatch('-')

plt.show()
```

```python
bars = plt.bar(l, vals, color='#FDD835')

bars[0].set_hatch('\\')
bars[1].set_hatch('/')
bars[2].set_hatch('//')
bars[3].set_hatch('x')
plt.show()
```

```python
fig = plt.figure()
patterns = [ "|" , "\\" , "/" , "+" , "-", ".", "*","x", "o", "O" ]

ax = fig.add_subplot(111)
for i in range(len(patterns)):
    ax.bar(i, 3, color='white', edgecolor='black', hatch=patterns[i])


plt.show()
```

```python
#different weight categories

light = fifa.loc[fifa.Weight < 60].count()[0] #will return count of all columns
light_medium = fifa.loc[(fifa.Weight >=60) & (fifa.Weight <70)].count()[0]

medium = fifa.loc[(fifa.Weight >= 70) & (fifa.Weight <=80)].count()[0]

med_heavy = fifa.loc[(fifa.Weight > 80) & (fifa.Weight <90)].count()[0]

heavy = fifa.loc[fifa.Weight>=90].count()[0]
```

```python
heavy
```

```python
med_heavy
```

```python
weights = [light, light_medium, medium, med_heavy, heavy]

labels = ['< 60', '60-70',  '70-80', '80-90', '> 90']

colors = ['#FF6384', '#36A2EB', '#FFCE56', '#8B008B', '#00FF00']
```

```python
plt.pie(weights, labels = labels, colors = colors)


plt.show()
```

```python
plt.style.use('ggplot')

plt.pie(weights, labels = labels, autopct = '%.2f %%')


plt.show()
```

**pctdistance=1.5**


*Percentage Distance from the center*

```python
plt.style.use('ggplot')

plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance=1.5)


plt.show()
```

##**Explode**

**explode = (.1, .1, .1, .1, .1)**

```python
explode = (.1, .1, .1, .1, .1)

plt.pie(weights, labels = labels, autopct = '%.2f %%', pctdistance=1.5, explode=explode)

plt.title('Weight Distribution of FIFA Players (kg)')

plt.show()
```

```python
plt.style.use('default')
```

##**Colors for each box**

```python
boxplot_colors = ['#ADD8E6', '#FFB6C1']

boxplot = plt.boxplot([barcelona, madrid], labels=['FC Barcelona', 'Real Madrid'], patch_artist=True)

# Set box colors
for box, color in zip(boxplot['boxes'], boxplot_colors):
    box.set(facecolor=color)

plt.title('Football Team Comparison')
plt.ylabel('FIFA Overall Ratings for Players')

plt.show()
```

```python
boxplot = plt.boxplot([barcelona, madrid], labels=['FC Barcelona', 'Real Madrid'])

cols = ['#4286f4', '#FFB6C1']

for box, i in zip(boxplot['boxes'], cols):
  box.set(color=i, linewidth=2)

plt.title('Football Team Comparison')
plt.ylabel('FIFA Overall Ratings for Players')


plt.show()
```

##**USING zip()**

```python
x = ['1', '2', '3']
h = ['A', 'X', 'Y']

for i,j in zip(x,h):
  print(j, ':>', i)
```

```python
boxplot = plt.boxplot([barcelona, madrid], labels=['FC Barcelona', 'Real Madrid'], patch_artist=True, medianprops={'linewidth':2})

colors = ['#4286f4', '#FFB6C1']

face_colors = ['#8dbdff', '#ffb6d9']

for box, i, j in zip(boxplot['boxes'], colors, face_colors):
  box.set(color=i, linewidth=2)
  box.set(facecolor = j)

plt.title('Football Team Comparison')
plt.ylabel('FIFA Overall Ratings for Players')


plt.show()
```
